## Connect CSVs


```python
# Specify the folder where your CSV files are located
folder_path = 'landmarks_csv'

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each CSV file and append its data to the list
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Save the combined DataFrame to a new CSV file if needed
combined_df.to_csv('combined_pose_landmarks.csv', index=False)

# Display the combined DataFrame
print(combined_df.head())
```

```python
pose_df = pd.read_csv('pole_pose_coords.csv')
pose_df
```

```python
pose_df['class name'].unique()
```

```python
## Check class distribution for imbalances

class_distribution = pose_df['class name'].value_counts()
print("class distribution")
print(class_distribution)
```

```python
from sklearn.model_selection import train_test_split
```

```python
shuffled_class_column = pose_df['class name'].sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
shuffled_class_column.value_counts()
```

```python
X = pose_df.drop('class name', axis=1) # features
y = shuffled_class_column # target value
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree
```

```python
## Check dimensions match
print(f'The shape of X_train is: {X_train.shape} and the shape of y_train is {y_train.shape}')
print(f'The shape of X_test is: {X_test.shape} and the shape of y_test is {y_test.shape}')
```

```python
pose_tree = DecisionTreeClassifier(criterion="entropy", max_depth=1)
pose_tree
```

```python
pose_tree.fit(X_train, y_train)
```

```python
#predict
pre_tree = pose_tree.predict(X_test)
print(pre_tree[0:5])
print(y_test[0:5])
```

```python
#Evaluate
from sklearn import metrics
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
dt_accuracy = metrics.accuracy_score(y_test, pre_tree)
print(f"DecisionTrees's Accuracy: {dt_accuracy}")
```
